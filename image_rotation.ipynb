```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as read
# from skimage.color import rgb2gray
raw_image=read.imread('cat.jpeg')
plt.imshow(raw_image)
plt.axis('off')
plt.show()
image=np.dot(raw_image[..., :3],[0.2989, 0.5870, 0.1140])
dimensions=image.shapeA
# print(dimensions)
image=np.transpose(image)
# print(image)
# taking the numbers of ccloums of the image
j=dimensions[1]
j-=1
for i in range(dimensions[1]):
    if i == j:
        break
    image[:, [i, j]] = image[:, [j, i]]
    j -= 1
plt.imshow(image, cmap='gray') 
# plt.imshow(image)
plt.axis('off')
plt.show()

# second way getting a grey coloured image
# gray_image=rgb2gray(raw_image)
# plt.imshow(gray_image)
# plt.axis('off')
# plt.show()
```
